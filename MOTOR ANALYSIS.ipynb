```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("colorado_motor_vehicle_sales.csv")

# View first 5 rows
df.head()
```

```python
# Shape of data
print("Rows & Columns:", df.shape)

# Column names
print(df.columns)

# Check missing values
print(df.isnull().sum())

# Data types
print(df.dtypes)
```

```python
df.describe()
```

```python
yearly_sales = df.groupby("year")["sales"].sum()

print(yearly_sales)


plt.figure()
yearly_sales.plot(kind="line")
plt.title("Year-wise Motor Vehicle Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()

```

```python
quarterly_sales = df.groupby("quarter")["sales"].sum()
print(quarterly_sales)


plt.figure()
quarterly_sales.plot(kind="bar")
plt.title("Quarter-wise Vehicle Sales")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.show()

```

```python
df["Year_Quarter"] = df["year"].astype(str) + "-Q" + df["quarter"].astype(str)

trend_sales = df.groupby("Year_Quarter")["sales"].sum()


plt.figure()
trend_sales.plot()
plt.title("Sales Trend by Year & Quarter")
plt.xlabel("Year-Quarter")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()

```
